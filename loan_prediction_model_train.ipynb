```python
X=df.drop(["Loan_Status"],axis=1)
Y=df[["Loan_Status"]]
```

```python
X.isnull().sum()
```

```python
X.info()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si1 = SimpleImputer(strategy='median')
si2 = SimpleImputer(strategy='most_frequent')
```

```python
col1=["LoanAmount","Loan_Amount_Term","Credit_History"]
col2=["Gender","Married","Dependents","Self_Employed"]
```

```python
for i in col1:
  X[[i]]=si1.fit_transform(X[[i]])
```

```python
for i in col2:
  X[[i]]=si2.fit_transform(X[[i]])
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
col=["Gender","Married","Dependents","Education","Self_Employed","Property_Area"]
```

```python
for i in col:
  X[i]=le.fit_transform(X[i].astype(str))
```

```python
X
```

```python
sns.heatmap(data=X)
```

```python
Y=le.fit_transform(Y.astype(str))
```

```python
Y
```

```python
sns.boxplot(X)
plt.xticks(rotation=90)
plt.show()
```

```python
from scipy.stats.mstats import winsorize
```

```python
w_col=["Gender","Dependents",'Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']
```

```python
for i in w_col:
  X[i]=winsorize(X[i],limits=[0.05,0.05])
```

```python
sns.boxplot(X)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,stratify=Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=500)
```

```python
lr.fit(X_train,Y_train)
lr_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print("********************************************")
print("Logistic Regression:",accuracy_score(Y_test,lr_pred))
print("Logistic Regression:",accuracy_score(Y_train,lr.predict(X_train)))
print("********************************************")
```

```python
print("**********************")
print("Logistic Regression:")
print(confusion_matrix(Y_test,lr_pred))
print("***********************")
```

```python
print("Logistic Regression:")
print(classification_report(Y_test,lr_pred))
```
